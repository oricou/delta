## Regrouper les données de `stats_generales`
Créer un fichier unique `data/stats_generales/stats_generales.csv` regroupe dans les données de `data/stats_generales`

```python
concat_file_list = []
for dirpath, dirname_list, filename_list in os.walk(STATS_GENERALES_PATH):
    print(dirpath)
    for filename in filename_list:
        if os.path.splitext(filename)[1] == ".tsv":
            print("ok")
            file_path = os.path.join(dirpath, filename)
            concat_file_list.append(file_path)
```

```python
len(concat_file_list)
```

```python
concat_file_list
```

```python
# initialise DataFrame pour la concat depuis une boucle
stats_generales_df = pd.read_csv(concat_file_list[0], delimiter="\t")
for file_path in concat_file_list[1:]:
    local_df = pd.read_csv(file_path, delimiter="\t")
    stats_generales_df = pd.concat([stats_generales_df, local_df], ignore_index=True)
```

```python
stats_generales_df
```

```python
stats_generales_export_path = os.path.join(STATS_GENERALES_PATH, "stats_generales.csv")
stats_generales_df.to_csv(stats_generales_export_path)
```

### Extraire les lignes regroupant les statistiques par banque
Comment les trouver ?
En regardant la présence de `¥` au début des noms dans `ecole` / qu'on peut vérifier par un regroupement par `banque` 

```python
stats_generales_df = pd.read_csv(STATS_GENERALES_FILE_PATH, index_col=0)
stats_generales_df
```

```python
# `ecole` avec NaN
stats_generales_df[stats_generales_df.ecole.isna()]
```

```python
# On va lui donner la valeur de la colonne `banque`
stats_generales_df.ecole.fillna(stats_generales_df[stats_generales_df.ecole.isna()].banque, inplace=True)
```

```python
# `ecole` avec NaN
stats_generales_df[stats_generales_df.ecole.isna()]
```

```python
# On extrait les résumés des stats par banque dont le nom d'école commence par `¥`
grouped_row_df = stats_generales_df.ecole.str.startswith("¥")
summary_stats_generales_df = stats_generales_df[grouped_row_df].copy()
# On enlève ces ligne de `stats_generales_df`
stats_generales_df.drop(summary_stats_generales_df.index, inplace=True)
summary_stats_generales_df.reset_index(inplace=True)
# On enlève le caractère `¥` de ce nom d'école
summary_stats_generales_df.ecole= summary_stats_generales_df.ecole.str.replace("¥", "")
summary_stats_generales_df
# Enregistrer les DataFrame
stats_generales_df.to_csv(STATS_GENERALES_FILE_PATH)
summary_stats_generales_df.to_csv(SUMMARY_STATS_GENERALES_FILE_PATH)
```

### Vérifier la cohérence des données

```python
summary_stats_generales_df = pd.read_csv(SUMMARY_STATS_GENERALES_FILE_PATH, index_col=0)
stats_generales_df = pd.read_csv(STATS_GENERALES_FILE_PATH, index_col=0)
```

```python
summary_stats_generales_df.head()
```

```python
summary_group = summary_stats_generales_df.groupby(by=["banque"]).sum()["integres_nb"]
summary_group
```

```python
general_group = stats_generales_df.groupby(by=["banque"]).sum()["integres_nb"]
general_group
```

On va étudier la liste des banques qui sont communes aux deux DataFrames (l'intersection). En effet, il n'y a qu'une quarantaine de ligne dans `summary_...` alors que l'autre en compte plus de 92 après le GROUPE BY. On peut facilement imaginer que c'est le fruit d'une fusion de plusieurs banques entre elles.
On voit que (même) dans ce sous-groupe, seules très peu de banques ont des chiffres cohérentes entre le  `summary_...` proposé par SCEI et un GROUP BY. Il peut s'agir de petites imprécisions entre les deux sources.

```python
intersection_banque = list(set(general_group.to_dict().keys()).intersection(set(summary_group.to_dict().keys())))
```

```python
general_group.eq(summary_group)[intersection_banque]
```

On a essaie de retrouver les données de `summary_stats_generales` en faisant un GROUP BY dans `stats_generales`.
Il y a certains concours comme `CONCOURS COMMUN INP ` et `CESI` où les nombres sont cohérents entre les deux tableaux. Mais pour le reste comme `EPITA` ou `A BIO`, les nombres sont complétements différents.
